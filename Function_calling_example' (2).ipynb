````python
import torch
from transformers import AutoProcessor, Gemma3ForConditionalGeneration
````

````python
model_id = "google/gemma-3-4b-it"

model = Gemma3ForConditionalGeneration.from_pretrained(
    model_id, device_map="auto"
).eval()

processor = AutoProcessor.from_pretrained(model_id)
````

````python
import requests

def get_weather(lat: float, lon: float):
    """
    Fetches live weather details for a given latitude and longitude using OpenWeatherMap API.

    Args:
        lat: The latitude of the location.
        lon: The longitude of the location.
    """
    API_KEY = API_KEY
    BASE_URL = "http://api.openweathermap.org/data/2.5/weather"

    params = {
        "lat": lat,
        "lon": lon,
        "appid": API_KEY,
        "units": "metric"  # Temperature in Celsius
    }

    try:
        response = requests.get(BASE_URL, params=params)
        response.raise_for_status()  # Raise an error for HTTP errors

        data = response.json()
        temp = data["main"]["temp"]
        humidity = data["main"]["humidity"]
        desc = data["weather"][0]["description"]

        return f"The weather at ({lat}, {lon}) is {desc} with a temperature of {temp}°C and humidity of {humidity}%."

    except requests.exceptions.RequestException as e:
        return f"Error fetching weather: {str(e)}"
````

````python
tools = [get_weather]
````

````python
query = "Can you tell me the weather at latitude 40.7128 and longitude -74.0060?"
````

````python
conversation = [
    {
        "role": "system",
        "content": [
            {
                "type": "text",
                "text": """
                You are an AI assistant that provides live weather updates using the OpenWeatherMap API.
                Use `get_weather(lat, lon)` when a user asks about weather for a specific location.

                - Always return a response based on the function output.
                - Ensure the function is wrapped in ```tool_code``` for execution.
                - Do not use `search()`, only use `get_weather(lat, lon)`.
                - If latitude and longitude are missing, ask the user for them.

                The following Python method is available:
                ```python
                def get_weather(lat: float, lon: float):
                    \"\"\"
                    Fetches weather data for a given latitude and longitude.

                    Args:
                       lat (float): Latitude of the location.
                       lon (float): Longitude of the location.

                    Returns:
                       dict: Weather details including temperature, conditions, etc.
                    \"\"\"
                ```

                User: {user_message}
                """
            }
        ]
    },
    {
        "role": "user",
        "content": [
            {"type": "text", "text": query}
        ]
    }
]
````

````python
inputs = processor.apply_chat_template(
            conversation,
            tools=tools,
            add_generation_prompt=True,
            return_dict=True,
            tokenize=True,
            return_tensors="pt",
).to(model.device, dtype=torch.bfloat16)
````

````python
import io
import re
import contextlib

def extract_tool_call(text):
    pattern = r"```tool_code\s*(.*?)\s*```"
    match = re.search(pattern, text, re.DOTALL)

    if match:
        code = match.group(1).strip()

        # Capture stdout to capture print statements
        stdout_buffer = io.StringIO()
        local_vars = {}

        try:
            with contextlib.redirect_stdout(stdout_buffer):
                exec(code, {}, local_vars)  # Execute code in a restricted scope

            # If there's a return value, extract it
            result = stdout_buffer.getvalue()
            if not result and "__result__" in local_vars:
                result = str(local_vars["__result__"])

            return f'```tool_output\n{result.strip()}\n```'
        except Exception as e:
            return f'Error executing tool code: {str(e)}'

    return None
````

````python
response = model.generate(**inputs, max_new_tokens=512)  # Ensure response exists
response_text = processor.decode(response[0], skip_special_tokens=True)

print("Raw Response:", response_text)  # Debugging output

keyword = extract_tool_call(response_text)
````

````python
prompt = [{
        "role": "user",
        "content": [
            {"type": "text", "text": f"Results from the search:{keyword} User_query : {query}"}
        ]
    }
]
````

````python
final_prompt = processor.apply_chat_template(
            prompt,
            tools=tools,
            add_generation_prompt=True,
            return_dict=True,
            tokenize=True,
            return_tensors="pt",
).to(model.device, dtype=torch.bfloat16)
````

````python
response = model.generate(**final_prompt, max_new_tokens=512)
````
